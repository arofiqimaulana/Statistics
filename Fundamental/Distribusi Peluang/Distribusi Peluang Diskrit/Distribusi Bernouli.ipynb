# 1. Distribusi Bernouli
Distribusi Bernoulli adalah model probabilitas untuk sebuah eksperimen yang hanya memiliki dua kemungkinan hasil: sukses (sering diberi label "1") dan gagal (sering diberi label "0"). Dalam konteks matematika, distribusi Bernoulli diwakili oleh parameter 
p, yang adalah probabilitas keberhasilan dalam satu percobaan. Dalam kehidupan sehari-hari, distribusi Bernouli dapat mewakili berbagai jenis eksperimen atau situasi. Berikut beberapa contoh:

## 1.1 Lemparan Koin
- Konteks: Ketika sebuah koin dilempar, ada dua kemungkinan hasil: angka atau gambar.

- Aplikasi: Probabilitas mendapatkan angka dapat diwakili sebagai p dan probabilitas mendapatkan sebagai 1-p.

## 2.1 Ujian
- Konteks: Seorang siswa menjawab satu pertanyaan benar/salah dalam ujian.

- Aplikasi: Probabilitas menjawab benar yaitu P(Benar) dapat diwakili sebagai p, sementara probabilitas menjawab salah P(Salah) dapat diwakili sebagai 1−p.

## 3.1 Kontrol Kualitas
- Konteks: Pemeriksaan item di jalur perakitan untuk melihat apakah produk cacat atau tidak.
- Aplikasi : Probabilitas menemukan produk yang cacat dapat diwakili oleh P(Cacat) = p. Sementara produk yang tidak cacat diwakili P(Tidak Cacat) = 1-p.

## 4.1 Pengambilan Sampel
- Konteks: Memilih satu item dari sekelompok item dan menentukan apakah item tersebut memiliki karakteristik tertentu atau tidak (mis., merah atau tidak merah).
- Aplikasi Distribusi Bernoulli: p dapat diwakili sebagai probabilitas memilih item yang merah, dan 1−p sebagai probabilitas item bukan merah.

## 5.1 Pengambilan Keputusan
- Konteks: Memutuskan untuk membeli atau tidak membeli produk setelah melihat iklan.
- Aplikasi: Probabilitas memutuskan untuk membeli P(Beli) dapat diwakili oleh p dan probabilitas memutuskan untuk tidak membeli P(Tidak Beli) dapat diwakili oleh 1−p.

## 6.1 Fenomena Alam
- Konteks: Prediksi cuaca misalnya, hujan atau tidak hujan pada hari tertentu.
- Aplikasi: Probabilitas hari itu hujan P(Hujan) dapat diwakili oleh p dan probabilitas hari itu tidak hujan P(Tidak Hujan)) dapat diwakili oleh 1-p


Di semua contoh di atas, kita melihat situasi di mana hanya ada dua kemungkinan hasil dan karenanya distribusi Bernoulli merupakan model yang tepat. Penting untuk diingat bahwa dalam prakteknya, **penentuan nilai *p* bisa didasarkan pada pengamatan historis, eksperimen, atau analisis lainnya**.

# 3. Graph

```python
from scipy.stats import bernoulli
import seaborn as sb

# Import the required libraries
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

# Instance of Bernoulli distribution with parameter p = 0.8
bd=bernoulli(0.8)

# Outcome of random variable either 0 or 1
x=[0,1]

# For the visualization of the bar plot of Bernoulli's distribution
plt.figure(figsize=(10,10))
plt.xlim(-2, 2)
plt.bar(x, bd.pmf(x), color='blue')

# For labelling of Bar plot
plt.title('Bernoulli distribution (p=0.8)', fontsize='20')
plt.xlabel('Values of random variable x (0, 1)', fontsize='20')
plt.ylabel('Probability', fontsize='20')

plt.show()
```

Sebuah dadu dilantunkan 1 kali. Jika diasumsikan sukses ketika muncul mata dadu genap. Tentukan
peluang sukses dari hasil lantunan sebuah dadu.

Solusi
- Dadu : {1,2,3,4,5,6}
- Dadu Genap : {2,4,6}

P(Dadu Genap) adalah

$$
\left(\frac{3}{6}\right)^1 \left(1-\frac{3}{6}\right)^{1-1} = (0.5)^1(0.5)^0 = 0.5
$$
